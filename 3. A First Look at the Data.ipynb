### Understanding the triaxial sensor layout

As a first step inspect the sensors information by plotting the labels of the first 40 sensors:

````python
ch_names = raw.info['ch_names']
print(ch_names[1:40])
````

The first 10 channels are trigger lines. Then follows 16 BNC channels which can be used to acquire digitized analogue input. The subsequent OPM sensors are organized according to triplets for each sensor. Each sensor measures the field in 3 directions denoted by X, Y and Z. The Z direction **CHECK** corresponds to conventional magnetometers measuring the field perpendicular to the scalp surface.  

For convenience we will here isolate the labels for the X, Y and Z directions:

````python
OPMX = [item for item in raw.ch_names if item.endswith("X")]
OPMY = [item for item in raw.ch_names if item.endswith("Y")]
OPMZ = [item for item in raw.ch_names if item.endswith("Z") and not item.startswith("BNC")]
````

In order to calculate on plot the power a few flat sensors must be set as bad:

````python
ch_names = raw.info['ch_names']
bad_chs = ['B4', 'H6']
bad_channels = [name for name in ch_names if any(ch in name for ch in bad_chs)]
raw.info['bads'] = bad_channels
raw.save
# Print the bad channels 
print("Bad channels marked:", raw.info['bads'])
````

### Estimating the power spectra
The data object (raw) allows for a simple inspection of the data by calculating the spectra: As a first step calculate the PSDs of all the OPM sensors for 0.1-120 HZ frequency range using the MNE functions. Also note that the sampling frequency 750 Hz and therefore chose $N_{FFT}$ = 1500 to achieve a 0.5 Hz spectral resolution.

````python
n_fft = 1500
raw_PSDX = raw.compute_psd(method="welch", fmin=0.1, fmax=120, picks=OPMX, n_fft=n_fft, n_overlap=int(n_fft/2))
raw_PSDY = raw.compute_psd(method="welch", fmin=0.1, fmax=120, picks=OPMY, n_fft=n_fft, n_overlap=int(n_fft/2))
raw_PSDZ = raw.compute_psd(method="welch", fmin=0.1, fmax=120, picks=OPMZ, n_fft=n_fft, n_overlap=int(n_fft/2))
````

To plot the PSDs for the OPMs for the X direction:

````python
%matplotlib inline  
raw_PSDX.plot(exclude='bads',picks='mag');
````

To plot the PSDs for the OPMs for the Y direction:

````python
%matplotlib inline  
raw_PSDY.plot(exclude='bads',picks='mag');
````

To plot the PSDs for the OPMs for the Y direction:

````python
%matplotlib inline  
raw_PSDZ.plot(exclude='bads',picks='mag');
````



Note the 50 Hz line noise. The frequency of the line noise is 50 Hz in European countries including Russia and Asian countries including India and China, It is 60 in South Korea as well as South and North America. In Japan it can be either 50 or 60 Hz depending on the region.
